## Background and context
For this project, we are interested in exploring any existing correlations between mental health disorder prevalence in specific countries and various factors including gross domestic product (GDP) per capita, pollution (CO2 emissions), population density, government style, and annual hours of sunshine. 

This research topic is broadly relevant, including in the US where mental health disorders have seemingly increased in prevalance dramatically in recent decades, leading to some even describing our country as experiencing a "mental health crisis." It is also especially relevant for Cornell students as mental health issues are a consistent and prevalent problem for our student body. A lot of college campuses experience mental health related problems, but Cornell is especially known for this and it has become a stereotype associated with our university. Therefore, our thinking behind designing our research question was that if we compare the rates of disorders in different countries by factors, we can make observations between correlations in disorder rates that could guide researchers concerned about this issue to the root causes of increased disorder rates.

We brainstormed and ultimately chose these specific factors because we thought they may have interesting connections to mental illnesses. GDP and eating disorders was a particularly interesting combination that is personally relevant for our group members; anecdotally, we have noticed that eating disorders seem to be more prevalent in wealthier communities, for example in one of our team member’s private high school eating disorders were unfortunately really common.  So, seeing if this trend is supported by our analysis and on a larger scale will be interesting. Population density and mental illness are another combination that interested us because we figured very low densely populated areas may have higher mental illness rates because of isolation and loneliness, but that there may be a threshold after which too high population density increases mental illness because of overcrowding and worse living conditions. A correlation CO2 emissions and mental illness intrigued us because of the simultaneous increase in mental illness and technology use and the associated global emissions in recent history. Government style and mental illness interested us because, from an American perspective at least, we view other types of governments as more restrictive and offering less freedom to citizens, so we wondered if this leads to worsened mental health. Lastly, annual hours of sunshine interested us because we’ve heard a lot about how important sunlight is for mental health with things like sun lamps growing in popularity to combat depression, so hypothesized that countries with fewer hours of sunlight would have higher rates of mental illness.

However, all of that being said, it is important to recognize from the start of our project that mental illness is a complex issue that can never be described by any single variable or piece of research. As we will also discuss at the conclusion of this project, there are inherent limitations to our data and analysis. # Part 2: Data Description

Each row of our analysis ready dataset represents a different country and its associated statistics. Each column of our dataframe represents a different aspect of the country, including its name, country code, prevalance of each disorder, and relevant statistics from each of the variables analyzed. There is therefore a column for each country's percentage of individuals with schizophrenia, depression, anxiety, bipolar disorder, eating disorders, and total disorder rate (sum of given percentages), as well the country's GDP per capita, population density, political regime, annual hours of sunshine, and CO<sub>2</sub> emissions per capita. We selected the most recent year in the data for the main analysis, 2019. We created this dataset by combining data from several different sources:

* The first, "Our World in Data," is a website whose mission is to provide “research and data to make progress against the world’s largest problems.” From this site, we retrieved data on mental illness prevalence by country, GDP per capita by country in international dollars based on purchasing power parity (PPP), and political regime by country. Based on their mission statement, it is assumed that this organization views mental health concerns, GDP, and political regimes as factors of some of the world’s largest problems globally. Our World in Data is funded by the charity Global Change Data Lab (GCDL), a registered charity in England and Wales, as well as by grants and direct website donations with donors listed on the site.
  
    * Our World in Data collected the disorder prevalence estimates through representative surveys, estimating the share of people with each mental illness in a given year, regardless of whether or not they were diagnosed. Therefore, it is likely that observed countries with fewer mental health stigmas and more open dialogues about mental health have more reported cases. Representative surveys were used, so it is assumed that individuals who completed the survey knew that they were doing so, but it is not explicitly stated whether or not they knew how the data would be used. During our data cleaning process, we also added our own column that summed the disorder percentages to get a total disorder prevalence statistic.
    * For the GDP data, Our World in Data compiled data from multiple sources by World Bank, including from their International Comparison Program (ICP) which is a large statistical initiative to compare world economies by converting volume and per capita measures of GDP into a common currency. The World Bank measures a country’s GDP by finding the sum of gross value added by all resident producers in the economy, plus any taxes, and minus any subsidies not included in product value. Most of this data comes from the statistical systems of member countries; therefore, one potential limitation is the possibility of a country misreporting its data for statistical analysis or having inconsistent methods for measuring these metrics.
      
    * The political regime data ranks government style of countries on a scale from 0 to 3: closed autocracies (0), electoral autocracy (1), electoral democracy (2), liberal democracy (3). The political regime dataset uses data from the V-Dem project which publishes data and research on democracy and human rights. The data is based on evaluations by “country experts” and supplemental work by project members to assess political systems. Therefore, this data is clearly at risk of being biased since it is based on subjective opinions on a country’s political system, even if they are “experts”. This is not objectifiable and measurable data in the same sense as other datasets. However, because of the large amount of experts coming to a consensus, it is likely more accurate. There is also little motive for impartial experts to fabricate information about countries’ political regimes. Here is some additional background on the political regime categories:
        * In closed autocracies, citizens do not have a say in choosing government leaders.
        * In electoral autocracies, citizens have a say in government leaders but are limited in some freedoms like expression or association which can make elections unfair and inaccessible.
        * In electoral democracies, citizens can choose government leaders in elections that are free and fair.
        * Liberal democracies are electoral democracies that include minority rights and equality for all citizens as well as the constraining of actions of government leaders by the legislature and courts.


* We also retrieved data on population density in people per square kilometer of land area and CO<sub>2</sub> emissions by country in metric tons per capita from The World Bank whose mission is “To create a world free of poverty on a livable planet.” This data was likely published to aid in this mission. The World Bank Group is funded by trust funds, core funding from fees paid by member countries, by member donations particularly from wealthier members, and from borrowing countries when they pay back loans. The population density data is mostly based on national population censuses, so a potential limitation of this data is the relatively high chance of miscounting population. The CO<sub>2</sub> emissions data is sourced from Climate Watch Historical GHG Emissions, which is production-based data meaning participating countries must submit emission inventories based on activities within their territories. Again, this opens the possibility of misreporting or variance in the quality of measurement tools that could be a limitation in the data. 

* The last factor in our dataframe is annual hours of sunlight per city in hours of sunshine per year, for which we retrieved data from Kaggle. Because the data was originally organized by city, we used the country columns in the dataset to average across the cities of each country for our final table. This is not an ideal metric for our project because within a given country, hours of sunlight can vary widely; for example, in the United States alone, there are northern cities that get very little sunlight during winter months, and cities closer to the equator that get many hours all year long. This was a large limitation in our project that potentially led to a lack of correlation in our eventual analysis. The source dataset was posted on Kaggle by someone who did the steps of web scraping a Wikipedia page to combine the data from many different tables. Kaggle is an online data science community with resources. There is definitely the limitation of the possibility of inaccurate data as anybody can post on Kaggle, especially in this case since data is sourced from Wikipedia. After analyzing the source Wikipedia page, however, the data seemed simple enough that we could trust it for this project.

## **Preregristration hypothesis 1:**
- $H_O$: Higher levels of GDP per capita are not correlated with higher levels of eating disorders, β = 0
- $H_A$: Higher levels of GDP per capita are correlated with higher levels of eating disorders, β > 0

We will reject the null hypothesis if our p-value is less than 0.05.

Here is a plot of the data examining eating disorder prevalence vs GDP per capita with a regression line:

```python
sns.regplot(x=df['GDP_per_capita'],y=df['Eating_percent']);
```

We must examine the residuals to see if any transformations are needed:

```python
sns.residplot(x=df['GDP_per_capita'],y=df['Eating_percent']);
```

This data looks like it might be heteroskedastic, so we used log transforms:

```python
df['log_GDP'] = np.log(df['GDP_per_capita'])
df['log_eating'] = np.log(df['Eating_percent'])
sns.residplot(x=df['log_GDP'],y=df['log_eating']);
```

This residual plot appears to be random so we will construct our model using the transformed columns. We will first perform cross validation to examine model performance and generalizability.

```python
X = df[['log_GDP']]
y = df['log_eating']
kf = KFold(n_splits = 5)
results = cross_validate(
    LinearRegression(),X,y,
    scoring='neg_root_mean_squared_error',
    cv=kf,return_train_score=True)
results_df = pd.DataFrame({
    'Train_RMSE':np.round(-1*results['train_score'],4),
    'Test_RMSE':np.round(-1*results['test_score'],4)})
print('Table summarizing train and test RMSE values across splits:') 
print(results_df)
print('Mean Train RMSE:',round(np.mean(results_df['Train_RMSE']),4))
print('Mean Test RMSE:', round(np.mean(results_df['Test_RMSE']),4))
print('Std Train RMSE:',round(np.std(results_df['Train_RMSE']),4))
print('Std Test RMSE:',round(np.std(results_df['Test_RMSE']),4))
print('Normalized Test RMSE:',round(
    np.mean(results_df['Test_RMSE'])/(df['log_eating'].max() - 
                                      df['log_eating'].min()),4))
```

Based on the RMSE values above, we can conclude: 
* The model seems to have consistent performance across different splits, as indicated by the similar train and test RMSE values for each split. The RMSE values are contained in a small range, from 0.3035-0.3736.
* The train RMSE values are relatively low. This indicates that in general the model fits the training data well. The mean test RMSE value of 0.326 suggests that on average, the model's predictions are within a reasonable range of the actual values in the test data.
* Looking across splits, the test RMSE values are relatively close to the training RMSE values. Additionally, the train and test RMSE means across splits are approximately equal, with the mean train RMSE at 0.3247 and the mean test RMSE at 0.3263. Thus, we can assume reasonable generalization to unseen data. 
* The standard deviation of the Train RMSE is 0.0066, showing relatively low variability in the model's predictions between training sets.
* The standard deviation of the Test RMSE is 0.0249, suggesting higher variability compared to the training set. Therefore, we can conclude that the model's ability to generalize to unseen data is less consistent. However, based on the fact that that the mean train RMSE and mean test RMSE are similar, and that the test standard deviation is still relatively low, we can still assume reasonable generalization.
* We decided to calculate and analyze the normalized RMSE value for hypothesis 1 to gain a better idea of if this RMSE value was “good.” Based on an article we found about normalizing RMSE, we calculated the normalized RMSE to get a value between 0 and 1 where values closer to 0 are considered “better” results. This is helpful because RMSE values are sometimes hard to compare because of unit inconsistencies and the range of the dataset itself. From this normalization, we got an RMSE value of 0.1044, which is relatively close to 0 and thus suggests the regression model fits the test data well. 
* The referenced article can be found at the following link: https://www.statology.org/what-is-a-good-rmse/#:~:text=Normalizing%20the%20RMSE%20Value&text=This%20produces%20a%20value%20between,is%20between%20%2470%2C000%20and%20%24300%2C000 

Through cross validation we have confirmed that a linear regression model works for this data. We now train an OLS linear regression model to look at the significance of our findings:

```python
X_with_constant = sm.add_constant(X)
model = sm.OLS(y,X_with_constant)
results = model.fit()
print(results.summary())
```

#### **Evaluation of significance:**
* **Log-GDP coefficient:** The `log_GDP` coefficient of 0.4510 and the fact that we performed a log-log transformation indicate that if a country's GDP per capita increases by 1%, the country's prevalance of eating disorders increases by 0.4510%. 
* **P-value of log-GDP:** The p-value for `log_GDP` is 0.000, which is less than 0.05, indicating that it is very unlikely that these results were achieved by chance so we reject the null hypothesis.  
* **Oddities/Outliers:** We expect this model to hold for `log_GDP` values between the minimum value from our data, 6 and a bit past the maximum value from our data, 12. This is because it is unrealistic to have a GDP per capita too low and we also cannot predict too far beyond our model's boundaries since we don't know if the linear relationship will always hold.

```python
def normalizer(vars):
    """Normalizes a distribution through z-scores"""
    return (vars - vars.mean())/vars.std()
```

## **Preregistration hypothesis 2:**
- $H_O$: Countries with liberal democracies have similar rates of eating disorders compared to countries that are not libreral democracies, $β_3$ = 0
- $H_A$: Countries with liberal democracies have higher rates of eating disorders compared to countries that are not liberal democracies, $β_3$ > 0 (and $β_3$ > $β_2$, $β_1$, $β_0$)

 We will reject the null hypothesis if our p-value is less than 0.05.

Below is a visualization for disorder prevalences per political regime:

```python
disorders = ['Depression_percent','Anxiety_percent',
             'Eating_percent','Schizophrenia_percent','Bipolar_percent']
for disorder in disorders:
    sns.barplot(df, x="Political_regime_score", 
                y=disorder)
    plt.show()
```

As shown in the bar graphs, a U-shape is a common pattern amongst disorder prevalences being compared to political regime. For all disorders except depression, we thought it was interesting that electoral autocracies (1) have the lowest disorder prevalence and liberal democracies (3) have the highest prevalence.

We will train a multivariable regression model to predict a country's eating disorder prevalence. First, we create dummy columns for the different political regimes:

```python
dummies = pd.get_dummies(
    df['Political_regime_score'],dtype=int)
regime_dummies = dummies.rename(
    columns={0:'is_closed_autocracy',
             1:'is_electoral_autocracy',
             2:'is_electoral_democracy',
             3:'is_liberal_democracy'})
df_with_dummies = pd.concat([df,regime_dummies],axis=1)
```

We create a correlation matrix to check for collinearity between variables, focusing on higher values:

```python
disorders = ['Schizophrenia_percent','Depression_percent',
             'Anxiety_percent','Bipolar_percent','Eating_percent']
regime_corr = df_with_dummies.drop(
    labels=['Entity','Code','Total_disorder_percent',
            'Political_regime_score']+disorders,axis=1).corr()
regime_corr[(regime_corr>0.5) | (regime_corr<-0.5)]
```

Since a country being a liberal democracy and GDP per capita appear to be correlated, we won't use GDP_per_capita and instead will use CO2_emissions as our additional variable.

We fit a multivariable linear regression model on all the data to check for randomness in residual plot:

```python
X = df_with_dummies[['CO2_emissions','is_electoral_autocracy',
                     'is_electoral_democracy', 
                     'is_liberal_democracy']]
y = df['Eating_percent']
multi_regime_model = LinearRegression().fit(X,y)
y_hat = multi_regime_model.predict(X)
residuals = y - y_hat
sns.scatterplot(x=y_hat,y=residuals)
plt.xlabel('Eating Disorder Percentage')
plt.ylabel('Residuals')
plt.axhline(0,linestyle=':');
```

The residuals look pretty random, so no transformations are needed. We will now perform cross validation to assess model performance and generalizability:

```python
kf = KFold(n_splits = 5)
results = cross_validate(
    LinearRegression(),X,y,
    scoring='neg_root_mean_squared_error',
    cv=kf,return_train_score=True)
results_df = pd.DataFrame({
    'Train_RMSE':np.round(-1*results['train_score'],4),
    'Test_RMSE':np.round(-1*results['test_score'],4)})
print('Table summarizing train and test RMSE values across splits:') 
print(results_df)
print('Mean Train RMSE:',round(np.mean(results_df['Train_RMSE']),4))
print('Mean Test RMSE:', round(np.mean(results_df['Test_RMSE']),4))
print('Std Train RMSE:',round(np.std(results_df['Train_RMSE']),4))
print('Std Test RMSE:',round(np.std(results_df['Test_RMSE']),4))
print('Normalized Mean Test RMSE:',round(
    np.mean(results_df['Test_RMSE'])/(df['Eating_percent'].max() - 
                                      df['Eating_percent'].min()),4))
```

Based on the RMSE values above, we can conclude:
* The train RMSE values are overall relatively low, indicating that the model fits the training data well. The Mean Train RMSE of 0.0959 suggests that on average the model's predictions are within a reasonable range.
* The test RMSE values are also overall relatively low and similar to the train RMSE values. The Mean Test RMSE of 0.0993 is also approximately equal to the mean train RMSE of 0.0959, meaning we can assume reasonable generalization to unseen data. 
* The standard deviation of the Train RMSE values is 0.0045, indicating a relatively low variability between training sets.
* The standard deviation of the Test RMSE values is 0.0169, showing slightly higher variability between the test sets than the training sets. # Part 7: Limitations



* A potential limitation of our data is an inherent part of one of the variables we are analyzing; countries with more authoritarian government regimes may not allow individuals to participate in the types of data collection used for these datasets and at the same time may under report mental health data. North Korea, for example, had low percentages of mental illness, but this may just be because of underreporting under the dictatorship.

* Another issue is the inherent variability in annual sunlight. Some countries (ex US, Russia, China) cover vast areas which leads to greater variability in hours of daylight. Within these larger countries there are also varying weather patterns (we know from experience Ithaca is cloudier than our hometowns!). To account for this, we may have to focus on smaller countries (set a square footage threshold) separately from these larger countries for the annual sunlight correlations.

* As mentioned previously, the original hours of sunlight dataset being organized by city instead of country was a large limitation that potentially led to a lack of correlation in this portion of the analysis. If we were to redo the project, we would need to find mental health data organized by city rather than country as well to better check for significance. From a broader perspective, this is a limitation to consider for other parts of our analysis; it is possible that analyzing entire countries is too broad for meaningful predictions about mental illness. Another approach to this project, if we were to continue in the future, could be analyzing statistics of different cities like the original sunlight hour dataframe or even college campus populations. There can be large variety within a given country between the cultures, enviornments, demographics, socioeconomic status, etc of different cities that may result in a more meaningful analysis of mental illness trends from a narrower scope.
